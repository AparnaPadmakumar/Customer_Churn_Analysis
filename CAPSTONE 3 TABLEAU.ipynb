```python
import pandas as pd
import matplotlib.pyplot as plt
from tabulate import tabulate   
import seaborn as sns
```

```python
df=pd.read_csv('DS1_C8_V3_ND_Sprint3_BankChurn_Data.csv')
df
```

```python
df.isnull().sum()               #count of Missing values in dataset
```

```python
df.info()                    # Count of non null values in dataset
```

```python
def seprate_data_types(df):
    categorical = []
    continuous = []
    for column in df.columns:                
        if df[column].nunique() < 30:       
            categorical.append(column)
        else:
            continuous.append(column)
            
    return categorical, continuous 
categorical, continuous = seprate_data_types(df)   
from tabulate import tabulate
table = [categorical, continuous]
print(tabulate({"categorical":categorical,
                "continuous": continuous}, headers = ["categorical", "continuous"]))
```

```python
#imputing values in card category
#column card category is categorical therefore replacing missing value by mode
card_category_mode = df.Card_Category.mode()[0]
df["Card_Category"].fillna(card_category_mode, inplace = True)
```

```python
#column Credit_Limit is continuous therefore replacing missing value by mean
Credit_Limit_mean = df.Credit_Limit.mean()
df["Credit_Limit"].fillna(Credit_Limit_mean, inplace = True)
```

```python
df.isnull().sum()
```

```python
#box plot for continuous columns
```

```python
for column in continuous:
    sns.boxplot(data=df[column])
    plt.title(f'Boxplot of {column}')
    plt.show()
```

```python
# By observing above box plot we can conclude that column Customer_Age, Dependent_count, Months_on_book, Credit_Limit, Avg_Open_To_Buy, Total_Trans_Amt and Total_Trans_Ct has outliers
```

```python
#Removing outliers by IQR method
```

```python
columns_with_outliers=['Customer_Age','Dependent_count','Months_on_book','Credit_Limit','Avg_Open_To_Buy','Total_Trans_Amt','Total_Trans_Ct']
for column in columns_with_outliers:
    Q1= df[column].quantile(0.25)
    Q3= df[column].quantile(0.75)
    IQR= Q3-Q1
    lower_bound= Q1-1.5*IQR
    upper_bound= Q3+1.5*IQR
    
    df=df[(df[column]>=lower_bound) & (df[column]<=upper_bound)]
```
